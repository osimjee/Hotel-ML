```python
#country doesn't want to convert for whatever reason so I'm going to remove it for the time being
hotel_data = hotel_data.drop(['country'], axis = 1)
hotel_data = hotel_data.drop(['agent'], axis = 1)
hotel_data = hotel_data.drop(['company'], axis = 1)
#agent and company have a ton of NA values
```

```python
from sklearn.model_selection import train_test_split
```

```python
#drop all rows that have na values
hotel_data = hotel_data.apply (pd.to_numeric, errors='coerce')
hotel_data = hotel_data.dropna()

print (hotel_data)
```

```python
#both of these come as floats and the model doesn't like that so switching them both to integers
hotel_data['children'] = pd.to_numeric(hotel_data['children'], downcast = 'integer')
hotel_data['adr'] = hotel_data['adr'].astype(int)
```

```python
data = hotel_data.drop(['reservation_status'],axis = 1)
```

```python
target = hotel_data['reservation_status']
```

```python
data_train, data_test, target_train, target_test = train_test_split(data,target, test_size = 0.30, random_state = 10)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gnb = GaussianNB()

pred = gnb.fit(data_train, target_train).predict(data_test)

print("Naive-Bayes accuracy : ",accuracy_score(target_test, pred, normalize = True))
```

```python
from sklearn.svm import LinearSVC
svc_model = LinearSVC(random_state=0, dual = False)
pred = svc_model.fit(data_train, target_train).predict(data_test)
print("LinearSVC accuracy : ",accuracy_score(target_test, pred, normalize = True))
```
